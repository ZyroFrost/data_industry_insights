```python
import pandas as pd
from pathlib import Path

# Define paths
ROOT = Path.cwd().parents[1]
INPUT_DIR = ROOT / "analysis" / "data" 
csv_path = INPUT_DIR / "final_jobs_500k.csv" # load global 500k data

csv_path2 = INPUT_DIR / "final_jobs_500k2.csv" # load global 500k data
print("csv_path: ", csv_path)

# Load data into DataFrame
df = pd.read_csv(
    csv_path,
    low_memory=False
)

df2 = pd.read_csv(
    csv_path2,
    low_memory=False
)

print(len(df))
print(len(df2))

missing_in_df2 = df.loc[~df["job_id"].isin(df2["job_id"])]
print("Missing in df2:", len(missing_in_df2))

missing_in_df = df2.loc[~df2["job_id"].isin(df["job_id"])]
print("Missing in df:", len(missing_in_df))

extra = df2.loc[~df2["job_id"].isin(df["job_id"])]

# đảm bảo posted_date là datetime
extra["posted_date"] = pd.to_datetime(extra["posted_date"], errors="coerce")

# tạo cột year từ posted_date
extra["year"] = extra["posted_date"].dt.year

# xem 20k này rơi vào năm nào
extra["year"].value_counts().sort_index()

#missing_in_df.to_csv("job_id_missing_in_df.csv", index=False)
```
